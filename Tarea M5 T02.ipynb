```python
from sklearn.preprocessing import RobustScaler #No normales con outlier
from sklearn.preprocessing import StandardScaler #Normales
from sklearn.preprocessing import MinMaxScaler #No normales sin outlier
```

```python
X_train_2= X_train.copy()
X_test_2= X_test.copy()
```

#### Aplicación de Prueba de Hipotesis Estadistica

```python
from scipy.stats import shapiro
for x in X_train_2.columns.values:
    stat, p = shapiro(X_train_2[x])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably Gaussian')
    else:
        print('Probably not Gaussian')
```

```python
X_train_2.columns.values[0]
```

```python
for x in X_test_2.columns.values:
    stat, p = shapiro(X_test_2[x])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably Gaussian')
    else:
        print('Probably not Gaussian')
```

Hacemos la normalizaciòn de datos, con el metodo shapiro para cada atributo del ocnjunto de datos para saber si son normales o no. 
el resultado es que no son normales.

#### EDA

Es importante identificar el tipo de variable o atributo que se tienen en el conjunto de datos, en este caso teniamos particularmente dos atributos de tipo categorico, por lo cual debemos procesar sus datos de diferente forma, el tipo de variable debe identificarse desde el principio para tener claro como preocesar los datos. 
para este caso tenemos dos variables categoricas ( CHAS Y RAD ), por lo cual no podemos estandarizar numericamente estos dos atributos:

Para la columna CHAS, aunque es categorica tambien en binaria por lo cual, no nos interesa transformarla.

Para columna RAD, existe mas de un valor, por lo cual se crean Dummys de la siguiente forma: 

```python
dum_train = pd.get_dummies(X_train_2.RAD, prefix="RAD_")
dum_train
```

```python
dum_test = pd.get_dummies(X_test_2.RAD, prefix="RAD_")
dum_test
```

Para columna RAD, se creo un Dummys tanto para Train como para Test, un Dummys permite convertir los valores contenidos en una columna en atributos y volverlos 0 y 1. (binarios)

```python
X_train_2=pd.concat([X_train_2, dum_train], axis=1)
X_train_2
```

```python
X_test_2=pd.concat([X_test_2, dum_test], axis=1)
X_test_2
```

Concatenamos Train + Dummis Train (67%) y Test + Dummis test (33%). con los datos ya transformados. 

```python
X_train_2=X_train_2.drop("RAD",axis=1)
```

```python
X_test_2=X_test_2.drop("RAD",axis=1)
X_test_2
```
