Es importante identificar el tipo de variable o atributo que se tienen en el conjunto de datos, en este caso teniamos particularmente dos atributos de tipo categorico, por lo cual debemos procesar sus datos de diferente forma, el tipo de variable debe identificarse desde el principio para tener claro como procesar los datos. 
para este caso tenemos dos variables categoricas ( CHAS Y RAD ), por lo cual no podemos estandarizar numericamente estos dos atributos:

Para la columna CHAS, aunque es categorica tambien en binaria por lo cual, no nos interesa transformarla.

Para columna RAD, existe mas de un valor, por lo cual se crean Dummys de la siguiente forma: 

```python
dum_train = pd.get_dummies(X_train_2.RAD, prefix="RAD_", drop_first=True)
dum_train
```

```python
dum_test = pd.get_dummies(X_test_2.RAD, prefix="RAD_", drop_first=True)
dum_test
```

Para columna RAD, se creo un Dummys tanto para Train como para Test, un Dummys permite convertir los valores contenidos en una columna en atributos y volverlos 0 y 1. (binarios), ademas con la funcion drop_first=True, evitamos la multicolinealidad, (correlacion entre columnas).

```python
X_train_2=pd.concat([X_train_2, dum_train], axis=1)
X_train_2
```

```python
X_test_2=pd.concat([X_test_2, dum_test], axis=1)
X_test_2
```
